```python
import numpy as np
```

```python
student_marks=np.array([[80,85,90],
                        [90,79,89],
                        [99,89,79],
                        [95,99,89],
                        [80,69,87]])
print(student_marks)
```

```python
print(student_marks.shape)
print(student_marks.ndim)
print(student_marks.size)
```

```python
avg_marks=np.mean(student_marks,axis=1)
print(f"Average marks of each student:{avg_marks.astype(int)}")
```

```python
min_marks=np.min(student_marks,axis=0)
print(f"Minimum marks of each subject:{min_marks}")
```

```python
max_marks=np.max(student_marks,axis=0)
print(f"Maximum marks of each subject:{max_marks}")
```

```python
topper=np.max(np.sum(student_marks,axis=1))
print(f"Topper of the class:{topper}")
```

```python
sum=np.sum(student_marks,axis=1)
print(f"Total marks of each student:{sum}")
```

```python
for i in range(len(avg_marks)):
    print(f"Average marks of student {i+1}:{avg_marks[i].astype(int)}")
    if (avg_marks[i] >= 90):
        print("Grade:A")
    elif (avg_marks[i] >=80):
        print("Grade:B")
    elif (avg_marks[i] >=70):
        print("Grade:C")
    else:
        print("Grade:D")
```
