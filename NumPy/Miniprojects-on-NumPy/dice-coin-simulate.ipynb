### Dice / Coin Simulation

**What to build**

- Simulate rolling dice or tossing coins 10,000 times
- Calculate probability of outcomes

**NumPy concepts**

- `np.random`
- Vectorized operations
- Frequency counting

```python
import numpy as np
```

```python
#Simulating rolling dice 10,000 times
rolling_dice=np.random.randint(1,7,10000)
print(f"Rolling dice 10,000 times:")
print(f"Outcomes : {rolling_dice}")
print(f"Length of outcomes :{len(rolling_dice)}")
#Calculating probability of outcomes
frequencies=np.bincount(rolling_dice)[1:]
probability=frequencies/10000
print(f"Calculating probability of outcomes:{probability}")
```
