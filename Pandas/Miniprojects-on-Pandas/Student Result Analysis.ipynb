```python
import pandas as pd
```

```python
student_marks = {
    "Names" : ["Alice", "Bob", "Charlie", "David", "Eve","Priya","Ria","Sriya","Mouni","Maanushi"],
    "Maths" : [25, 92, 78, 96, 99, 90, 82, 40,98,99],
    "CS": [38, 90, 45, 60, 98, 99, 62, 78, 89, 92],
    "Stats" : [98, 96, 89, 76, 98, 97, 89, 67, 79 ,86]
}
df1 = pd.DataFrame(student_marks)
df1
```

```python
df = df1.to_csv("student_marks.csv", index=False)
```

```python
df = pd.read_csv("student_marks.csv")
```

```python
#Calculate average marks
average_marks = df[["Maths","CS","Stats"]].values.mean()
print(f"Average marks of a students: {average_marks}")
```

```python
#Find topper
df["total"] = df[["Maths","CS","Stats"]].sum(axis=1)
topper_name = df.loc[df["total"].idxmax()]
print("Topper:", topper_name)

```

```python
#Find subject-wise highest score
high_score = df[["Maths","CS","Stats"]].max(axis=0)
print("subject-wise highest score:")
print(high_score)
```

```python
#Filter students who failed
failed_students = df[
    (df[["Maths", "CS", "Stats"]] <= 40).any(axis=1)
]

print(failed_students)
```
