# Rows and Columns - Operation (Add, update, delete)

```python
# Create a new column
df['Team']=['CEO', 'HR', 'CTO', 'DA', 'SDE']
df
```

```python
# Add new columns using existing column/s
df['Bonus']=df['monthly_salary']*1.5
df
```

```python
df.loc[len(df)]=['sus',25,34000,'SDE1',(34000*1.5)] # adding new row to dataframe
```

```python
df.drop(df[df.Name=='sus'].index,inplace=True) # deleting a row from dataframe
```

```python
df
```

```python
#Update a value using index in dataframe
df.loc[0,'monthly_salary'] = 100000
df
```

```python
#update value using column name
df.loc[df.Name=='priya','monthly_salary']=95000
df
```

```python
#dropping a column
df.drop('Bonus',axis=0,inplace=True)
#del multiple columns
#df.drop(['Bonus','Team'],axis=0,inplace=True)
```

```python
df.drop(1,axis=0) #can use inplace=True to make changes permanent
```
