```python
import matplotlib.pyplot as plt
```

```python
#create first chart
x=[1,2,3,4,5,6]
y=[10,25,30,5,40,80]
#plotting x,y in chart
plt.plot(x,y)  #line chart
#it will show chart 
plt.show()
```

```python
plt.figure(figsize=(6,5))  #figure size
plt.plot(x,y,color= 'purple' ,marker='.',linestyle='dashed',linewidth=2,markersize=12)
plt.title("Sample Line Chart")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```

```python
x=[1,2,3,4,5,6]
y1=[10,25,30,5,40,80]
y2=[15,40,35,15,45,85]
plt.plot(x,y1,label='Sales 2024',color="green",marker='o',linestyle='dotted')
plt.plot(x,y2,label='Sales 2025',color="red",marker='o',linestyle='dotted')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.title("Sales Information in the years 2024 and 2025")
plt.legend()
plt.show()
```

```python
plt.bar(x,y)
plt.title("Barchart Example")
plt.show()
```

```python
# Histogram
#used for distribution analysis
import random
data=[random.randint(1,10) for _ in range(100)]

plt.hist(data,bins=10)
plt.title("Histogram Example")
plt.show()
```

```python
#pie chart
x=['A','B','C','D','E']
y=[20,50,90,75,80]

plt.pie(y,labels=x,autopct='%1.1f%%',startangle=45)
plt.title("Example of Piechart")
plt.show()
```

```python
#scatter plot
x=['A','B','C','D','E']
y=[10,25,30,40,80]

x1=[1,2,3,4,5,6]
y1=[15,40,35,15,45,85]

plt.figure(figsize=(10,5))  

plt.subplot(1,2,1) #row,column,position
plt.bar(x,y)
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")




plt.subplot(1,2,2) #row,column,position
plt.scatter(x1,y1)
plt.title("Example of Scatter")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()

```

### Matplotlib with Pandas - real data

```python
# create df
import pandas as pd

# data
data = {
    'Month' : ['Jan', 'Feb', 'Mar', 'Apr'],
    'Sales' : [12000, 11000, 13000, 25000]
}

df = pd.DataFrame(data)
df
```

```python
plt.bar(df['Month'], df['Sales'])
plt.title("Matplotlib with Pandas")
plt.xlabel("Month")
plt.ylabel("Sales")

plt.savefig("Monthly_Sales_from_df.png")

plt.show()
```
